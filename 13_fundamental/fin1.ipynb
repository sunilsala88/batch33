```python
dict_of_beta={}
list_of_stocks=['AAPL','MSFT','GOOGL','AMZN','TSLA','META','NVDA','JPM','V','UNH','HD','PG','MA','DIS','PYPL','BAC','ADBE','CMCSA','NFLX','XOM','KO']
for stock in list_of_stocks:
    t1=yf.Ticker(stock)
    i1=t1.info
    beta=i1.get('beta')
    dict_of_beta[stock]=beta
```

```python
l1=list(dict_of_beta.values())
i=l1.index(max(l1))
list_of_stocks[i]
```
