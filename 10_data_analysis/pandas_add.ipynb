```python
import pandas as pd
import numpy as np

np1=np.arange(1,26).reshape(5,5)
df1=pd.DataFrame(np1,columns=['c1','c2','c3','c4','c5'],index=['r1','r2','r3','r4','r5'])
df1.index.name='c0'

np2=np.arange(101,126).reshape(5,5)
df2=pd.DataFrame(np2,columns=['c6','c7','c8','c9','c10'],index=['r1','r2','r3','r4','r5'])
df2.index.name='c0'
```

```python
print(df1)
df2
```

```python
pd.merge(df1,df2,on='c0')
```
